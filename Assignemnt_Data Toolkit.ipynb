**Q2.How does broadcasting work in NumPy?**

Ans:Broadcasting in NumPy refers to the ability to perform element-wise operations on arrays of different shapes in a way that NumPy automatically adjusts the smaller array to match the shape of the larger one, without needing explicit replication of data.

How Broadcasting Works:

**Shape Comparison:**

NumPy compares the shapes of the two arrays element-wise, starting from the trailing dimensions (the rightmost ones).
If the dimensions are different, NumPy tries to "stretch" the smaller array to match the size of the larger array by repeating its values along the missing dimensions, if possible.

**Rules of Broadcasting**:

Rule 1: If the arrays have a different number of dimensions, NumPy pads the smaller array’s shape with ones on the left side (until both shapes have the same length).
Rule 2: Two dimensions are compatible if they are equal, or one of them is 1.
Rule 3: If a dimension of one array is 1 and the other array’s dimension is greater than 1, the smaller array will be "stretched" along that dimension.

**Q4.Explain the use of the groupby() method in Pandas?**

Ans: The groupby() method in Pandas is used to split data into groups based on some criteria (such as values in one or more columns), perform operations on each group independently, and then combine the results. It’s commonly used for tasks like aggregation, summarization, and transformation.

**Use Cases:**

**Aggregation:** Sum, mean, count, etc., for each group.

**Transformation:** Apply functions to modify the data within groups.

**Filtering:** Select groups based on certain conditions.

```python
import pandas as pd

# Sample DataFrame
data = {'Category': ['A', 'A', 'B', 'B', 'A', 'B'],
        'Value': [10, 20, 30, 40, 50, 60]}

df = pd.DataFrame(data)

# Grouping by 'Category' and calculating the sum of 'Value' for each group
grouped = df.groupby('Category')['Value'].sum()
print(grouped)
```

**Q20.What is the role of a pivot table in Pandas?**

Ans: A pivot table in Pandas is used to summarize and aggregate data in a DataFrame, allowing you to reorganize and analyze data in a more structured way.

Role of Pivot Table:
Data Aggregation: It allows you to group data by one or more columns and apply aggregate functions (e.g., sum, mean, count) to summarize the data.
Reshaping Data: It reshapes data by turning unique values from one column into separate columns, making the data easier to analyze.
Multi-dimensional Analysis: Supports multi-level grouping, helping you analyze data from multiple perspectives.

```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Sales': [100, 200, 150, 250, 300]
})

# Creating a pivot table
pivot_table = pd.pivot_table(df, values='Sales', index='Category', aggfunc='sum')
print(pivot_table)
```

**Q21.Why is NumPy's array slicing faster than Python's list slicing?**

Ans:**NumPy's array slicing** is faster than **Python's list slicing** due to the following reasons:

1. **Contiguous Memory**: NumPy arrays are stored in **contiguous memory locations**, allowing efficient access and slicing without the overhead of Python's dynamic list structure.

2. **Optimized C Implementation**: NumPy operations are implemented in **C**, which is much faster than Python's loop-based list slicing.

3. **No Copying**: NumPy slicing often creates **views** (instead of copies) of the original array, making slicing operations faster and more memory-efficient. Python lists, however, create a new copy of the sliced data.

**Q4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [5, 6, 7, 8, 9],
        'C': [9, 8, 7, 6, 5],
        'D': [2, 4, 6, 8, 10]}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

# Display the plot
plt.title('Correlation Matrix Heatmap')
plt.show()
```
